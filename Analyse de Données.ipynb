```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as stat
import pandas as pd
```

```python
x = [19, 60, 30, 76, 48]
y = [21.62, 40, 50, 69, 40]
plt.scatter(x, y, c='r')
plt.show()
plt.grid()
```

```python
a, b, r, p, ES = stat.linregress(x, y)
print("a=",a, "\nb=",b, "\nr=",r, "\nPvalue=",p, "\nErreur std=", ES)
```

```python
def droit(x):
    return a*x + b
M = list(map(droit, x))
M
```

```python
plt.scatter(x, y, c='r')
plt.plot(x, M, c='b')
```

```python
x = np.array(x)
y = np.array(y)

plt.scatter(x, y, c='r')
plt.plot(x, a*x+b, c='b')
```

```python
# linspace utilisé pour les fonctions
M = np.linspace(-5, 5, 40)
plt.plot(M, M**2, label="y=X²")
plt.plot(M, M**3, label="z=X^3")
plt.title("Les fonctions avec linspace")
plt.xlabel("X")
plt.ylabel("Y=X² et z=X^3")
plt.legend()
plt.show()
```

# Test d'independance de KHI-DEUX X²

```python
M = np.array([[16, 18, 12, 10], [18, 20, 24, 27,], [12, 16, 13, 14]])
M
```

```python
from scipy.stats import chi2_contingency as kdeux
```

```python
d, p, ddl, tt = kdeux(M)
print("d=",d, "\nddl=",ddl, "\nPvalue=",p)
print("\nTavleau théorique=\n", tt)
```
