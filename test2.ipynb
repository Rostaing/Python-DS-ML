```python
bitcon = pd.read_csv('BTC-EUR.csv', index_col='Date', parse_dates=True)
```

```python
bitcon.head()
```

```python
bitcon.columns
```

```python
bitcon['Close'].plot(figsize=(9, 6))
plt.show()
```

```python
bitcon['2019-09']['Close'].plot()
```

```python
bitcon.loc['2017':'2019', 'Close'].plot()
```

```python
bitcon.loc['2019', 'Close'].resample('M').plot()
plt.show()
```

```python
bitcon.loc['2019', 'Close'].resample('M').mean().plot()
plt.show()
```

```python
bitcon.loc['2019', 'Close'].resample('2W').plot()
plt.show()
```

```python
bitcon.loc['2019', 'Close'].resample('M').agg(['std', 'max', 'min']).plot(figsize=(9, 6))
plt.show()
```

```python
bitcon.loc['2019', 'Close'].rolling(window=7, center=True).mean().plot()
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.drop(['alone', 'alive', 'who', 'adult_male', 'embark_town', 'class'], axis=1, inplace=True)
```

```python
titanic.dropna(axis=0, inplace=True)
```

```python
titanic.head()
```

```python
sns.pairplot(titanic)
```

```python
sns.catplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
```
